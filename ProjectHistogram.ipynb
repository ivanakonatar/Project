

 # Histogram


```python
import matplotlib.pyplot as plt 
from scipy.misc import imread,imsave
%matplotlib inline

rgb_image = imread("a.jpg")
plt.imshow(rgb_image)


```

```python
opseg = range(0, 256)

def f(color):
    d = {}
    for i in range(len(color)):
        for j in range(len(color[0])):

            if color[i][j] not in d:

                y = 0

                for m in range(i, len(color)):
                    for n in range(j, len(color[0])):

                        if color[i][j] == color[m][n]:
                            y += 1

                d[color[i][j]] = y
    for k in opseg:
        if k not in d:
            d[k] = 0
    return dict(sorted(d.items()))
# print(d)


```

## Histogram RED


```python
red_image = rgb_image[:, :, 0]
d_r = f(red_image)


plt.bar(range(len(d_r)), list(d_r.values()), align='center', color='red')
plt.xticks(range(len(d_r)), list(d_r.keys()))

plt.show()

```

## Histogram GREEN

```python
green_image = rgb_image[:, :, 1]
d_g = f(green_image)

plt.bar(range(len(d_g)), list(d_g.values()), align='center', color='green')
plt.xticks(range(len(d_g)), list(d_g.keys()))

plt.show()
```

## Histogram BLUE

```python
blue_image = rgb_image[:, :, 2]
d_b = f(blue_image)

plt.bar(range(len(d_b)), list(d_b.values()), align='center', color='blue')
plt.xticks(range(len(d_b)), list(d_b.keys()))

plt.show()
```

## Histogram RGB

```python
plt.bar(range(len(d_r)), list(d_r.values()), align='center', color='red')
plt.xticks(range(len(d_r)), list(d_r.keys()))
plt.bar(range(len(d_g)), list(d_g.values()), align='center', color='green')
plt.xticks(range(len(d_g)), list(d_g.keys()))
plt.bar(range(len(d_b)), list(d_b.values()), align='center', color='blue')
plt.xticks(range(len(d_b)), list(d_b.keys()))

plt.show()
```
